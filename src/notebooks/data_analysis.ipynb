```python
# Libraries
import os

import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
# Load the Iris dataset from sklearn
iris = datasets.load_iris()
print(iris)
```

```python
# Convert the dataset into a DataFrame
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['species'] = iris.target
display(df_iris)
```

```python
# Save original dataset
df_iris.to_csv('../data/iris.csv', index=False)
```

```python
# Split the dataset and save the sets to a .CSV
iris_train, iris_validation = train_test_split(df_iris, test_size=0.2, random_state=42)
iris_train.to_csv('../data/iris_train.csv', index=False)
iris_validation.to_csv('../data/iris_validation.csv', index=False)

print("The files iris_train.csv and iris_validation.csv have been saved.")
```

```python
# Delete extra files
files_to_delete = ['../data/iris_train_sample.csv', 
                   '../data/iris_train.csv', 
                   '../data/iris_validation.csv',
                   '../data/iris.csv']

for file in files_to_delete:
    if os.path.exists(file):
        os.remove(file)
        print(f"{file} deleted")
    else:
        print(f"{file} does not exist")
```
